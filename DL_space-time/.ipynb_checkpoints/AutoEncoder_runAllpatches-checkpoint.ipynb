```python
from DeepLearnFuncs import *
import six.moves.cPickle as cPickle

import matplotlib.pyplot as plt
%matplotlib inline  
plt.rcParams['figure.figsize'] = (16, 10)
```

```python
# start by importing Deep Learning Funcs
funcs = DeepLearnFuncs()

learning_rate=0.1
training_epochs=20
corruptionL=0.0
batch_size=20
output_folder='dA_plots'

############
# train dAutoencoder                 
############
dfLLdata = funcs.test_dA(learning_rate, training_epochs, corruptionL, batch_size, output_folder)
```

```python
 # change correpution level to 0.35
corruptionL=0.35

############
# train dAutoencoder                 
############
dfLLdata = funcs.test_dA(learning_rate, training_epochs, corruptionL, batch_size, output_folder)

############
### plotting or cost
### the cost we minimize during training is the negative log likelihood of
############
x = dfLLdata['iter'].values
y = dfLLdata['LL_iter'].values
plt.figure()
plt.plot(x, y, 'bo--')
plt.xlabel('epoch', fontsize=14)
plt.ylabel('dA Cost', fontsize=14)
plt.title('MLP: learning_rate = '+str(learning_rate)+' batch_size = '+str(batch_size), fontsize=14)
```

```python
 # change correpution level to 0.10 but decrease learning rate and increase learning epochs
corruptionL=0.1
learning_rate=0.025
training_epochs=100


############
# train dAutoencoder                 
############
dfLLdata = funcs.test_dA(learning_rate, training_epochs, corruptionL, batch_size, output_folder)

############
### plotting or cost
### the cost we minimize during training is the negative log likelihood of
############
x = dfLLdata['iter'].values
y = dfLLdata['LL_iter'].values
plt.figure()
plt.plot(x, y, 'bo--')
plt.xlabel('epoch', fontsize=14)
plt.ylabel('dA Cost', fontsize=14)
plt.title('MLP: learning_rate = '+str(learning_rate)+' batch_size = '+str(batch_size), fontsize=14)
```
